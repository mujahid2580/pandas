# THE LEFT_ON AND RIGHT_ON

```python
df3= pd.DataFrame({'name':['raju','rani','raghu','ram','ramesh'],
                  'salary':[70000,80000,50000,60000,40000]})
pd.merge(df1,df3,left_on='employee',right_on='name')
```

```python
pd.merge(df1,df3,left_on='employee',right_on='name').drop('name',axis=1)
```

# HOW TO PERFORM SET ARITHMATIC JOINS

```python
df6 = pd.DataFrame({'name':['peter','mary','paul'],
                   'food':['fish','beans','bread']},
                  columns=['name','food'])
df7 = pd.DataFrame({'name':['mary','paul','joseph'],
                   'drink':['wine','beer','water']},
                  columns=['name','drink'])
```

```python
pd.merge(df6,df7)
```

```python
pd.merge(df6,df7,how='inner')
```

```python
pd.merge(df6,df7,how='outer')
```

```python
pd.merge(df6,df7,how='left')
```

```python
pd.merge(df6,df7,how='right')
```

# OVERLAPING COLUMN NAMES: THE SUFFIX KEYWORD

```python
df8 = pd.DataFrame({'name':['bob','jake','lisa','sue'],
                   'rank':[1,2,3,4]})
df9 = pd.DataFrame({'name':['bob','jake','lisa','sue'],
                   'rank':[3,1,4,2]})
pd.merge(df8,df9,on="name")
```

```python
pd.merge(df8,df9,on="name",suffixes= ["_l","_r"])
```
