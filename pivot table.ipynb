# PIVOT TABLE SYNTAX

```python
titanic.pivot_table('survived',columns='sex')
```

# MULTI LEVEL PIVOT TABLE

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

#### here cut is used to cut the data according to ages  cut is used for time series analysis

```python
age = pd.cut(titanic['age'],[0,18,30,80])
titanic.pivot_table('survived',['sex',age,'who'],'class')
```

#### qcut is used for quantile statistics which is measured

```python
fare = pd.qcut(titanic['fare'], 4)
titanic.pivot_table('survived',['sex',age,'who'],[fare,'class'])
```

## additional pivot table

```python
titanic.pivot_table(index='sex',columns='class',
                   aggfunc=({'survived':sum,'fare':'min'}))
```
