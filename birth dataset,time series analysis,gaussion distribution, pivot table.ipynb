```python
b['births'].median()
```

```python
quartiles = np.percentile(b['births'],[25,50,75])
quartiles
```

```python
mu = quartiles[1] # here 1 in dex of quartiles
mu
```

```python
sig = 0.74 * (quartiles[2]-quartiles[0]) # 
sig
```

```python
births = b.query('(births > @mu - 5 * @sig & births < @mu +5 * @sig)')  # m-a*sigma , m+a*sigma
births
```

```python
births['day']= births['day'].astype(int)
```

```python
births.head()
```

```python
births.isnull().describe()
```

```python
births.isna().sum()
```

```python
# create a datetime index from the year month and day
births.index = pd.to_datetime(10000*births.year +
                          100*births.month +
                          births.day,format='%Y%m%d')
births['day_of_the_week']= births.index.dayofweek
births
```

```python
# gca is used to set the xlabel 
births.pivot_table('births',index='day_of_the_week',
                    columns='decade',aggfunc='mean').plot()
plt.gca().set_xticklabels(['mon','tue','wed','thur','fri','sat','sun'])
plt.ylabel('mean births by day')
```

```python
births_by_date = births.pivot_table('births',
                                   [births.index.month,births.index.day])
births_by_date
```

```python
# for particular year
births_by_date.index = [pd.datetime(2012, month ,day)
                       for (month,day) in births_by_date.index]
births_by_date.head()
```
