# HERE iloc AND loc IS USED loc gets rows (or columns) with particular labels from the index. iloc gets rows (or columns) at particular positions in the index(so it takes only integers)

```python
frame = pd.DataFrame(np.random.randn(1000, 5),   columns=['a', 'b', 'c', 'd', 'e'])
frame.iloc[::2] = np.nan
frame
```

```python
frame['f']=np.random.randn(1000) # to add the column name
frame
```

```python
frame['g']=np.random.randn(1000)
frame
```
